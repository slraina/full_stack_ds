**1. Write a Python Program to Check if a Number is Positive, Negative or Zero?**

```python
num = int(input("Enter a number: "))

if num>0:
    print('\nThis is a positive number.')
elif num<0:
    print('\nThis is a negative number.')
else :
    print('\nThis is Zero.')
```

**2. Write a Python Program to Check if a Number is Odd or Even?**

```python
num = int(input("Enter a number: "))

if num%2==0:
    print('\nThis is a even number.')
else :
    print('\nThis is odd number.')
```

**3. Write a Python Program to Check Leap Year?**

```python
yr = int(input("Enter the year: "))

if (yr % 100 == 0) and (yr%400 == 0):
    print(f"{yr} is a leap year")
elif (yr%100 !=0)and (yr%4 == 0):
    print(f"{yr} is a leap year")
else : 
    print(f"{yr} isn't a leap year.")
```

**4. Write a Python Program to Check Prime Number?**

```python
import math as ma
num = int(input("Enter the year: "))

ns = ma.ceil(ma.sqrt(num))+1

if num == 1:
    print("Not Prime")
elif num == 2:
    print("Prime Number")
else:
    for i in range(2,ns):
        if (num%i == 0):
            print(f"\nNumber {num} isn't a prime number.")
            break
    else:
        print("\nPrime number")
```
